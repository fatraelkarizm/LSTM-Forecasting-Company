```python
import matplotlib.pyplot as plt
import json

# 1. Baca data hasil prediksi
with open('outputs/forecast_output.json', 'r') as f:
    data = json.load(f)

categories = []
amounts = []
colors = []

# 2. Ambil data Inflow (Hijau)
for item in data['inflow']:
    categories.append(item['category'])
    amounts.append(item['predicted_amount'])
    colors.append('#2ecc71') # Warna Hijau Keren

# 3. Ambil data Outflow (Merah)
for item in data['outflow']:
    categories.append(item['category'])
    amounts.append(item['predicted_amount'])
    colors.append('#e74c3c') # Warna Merah Keren

# 4. Bikin Grafik
plt.figure(figsize=(12, 6)) # Ukuran gambar
bars = plt.bar(categories, amounts, color=colors)

# Tambahkan angka di atas batang biar jelas
for bar in bars:
    yval = bar.get_height()
    # Format angka jadi format uang (pakai koma)
    formatted_num = f"{int(yval):,}"
    plt.text(bar.get_x() + bar.get_width()/2, yval, formatted_num, va='bottom', ha='center', fontweight='bold')

# Hiasan Judul & Label
total_in = int(data['predicted_inflow'])
total_out = int(data['predicted_outflow'])
plt.title(f"PREDIKSI CASHFLOW 30 HARI KE DEPAN\nTotal Masuk: Rp {total_in:,}  |  Total Keluar: Rp {total_out:,}", fontsize=14, fontweight='bold')
plt.ylabel("Nominal (Rupiah)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Tampilkan
plt.show()
```
